```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/content/imageconversion.jpg')
```

```python
plt.imshow(img)
```

```python
grayimg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grayimg)
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(16, 16))
image = clahe.apply(grayimg) 
```

```python
plt.imshow(image)
```

```python
color = cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
```

```python
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(color)
```
