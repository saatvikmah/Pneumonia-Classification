```python
import pandas as pd
```

```python
inf_Inception = pd.read_csv('/content/drive/MyDrive/classification/saved Models/Pretrained InceptionV3/predictions.csv')
inf_VGG = pd.read_csv('/content/drive/MyDrive/classification/saved Models/Pretrained VGG16/predictions.csv')
inf_Dense = pd.read_csv('/content/drive/MyDrive/classification/saved Models/Pretrained DenseNet121/predictions.csv')
inf_Resnet = pd.read_csv('/content/drive/MyDrive/classification/saved Models/Pretrained ResNet50/predictions.csv')
inf_Inception1 = pd.read_csv('/content/drive/MyDrive/classification/saved Models/Pretrained InceptionV3_1/predictions.csv')
```

```python
from matplotlib.figure import Figure
import sklearn.metrics as metrics
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,6)
fpr, tpr, thresholds = metrics.roc_curve(inf_Inception.label, inf_Inception.pred_prob)
roc_auc = metrics.auc(fpr, tpr)
plt.plot(fpr, tpr, 'b', label = 'InceptionV3 (512X512) = %0.2f' % roc_auc, color='red')

fpr, tpr, thresholds = metrics.roc_curve(inf_VGG.label, inf_VGG.pred_prob)
roc_auc = metrics.auc(fpr, tpr)
plt.plot(fpr, tpr, 'b', label = 'VGG = %0.2f' % roc_auc, color = 'green')

fpr, tpr, thresholds = metrics.roc_curve(inf_Dense.label, inf_Dense.pred_prob)
roc_auc = metrics.auc(fpr, tpr)
plt.plot(fpr, tpr, 'b', label = 'DenseNet121 = %0.2f' % roc_auc, color='blue')

fpr, tpr, thresholds = metrics.roc_curve(inf_Resnet.label, inf_Resnet.pred_prob)
roc_auc = metrics.auc(fpr, tpr)
plt.plot(fpr, tpr, 'b', label = 'ResNet50 = %0.2f' % roc_auc, color='black')

fpr, tpr, thresholds = metrics.roc_curve(inf_Inception1.label, inf_Inception1.pred_prob)
roc_auc = metrics.auc(fpr, tpr)
plt.plot(fpr, tpr, 'b', label = 'InceptionV3 (224x224) = %0.2f' % roc_auc, color='orange')

plt.legend(loc = 'lower right', title = 'AUC Score')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.show()
```
